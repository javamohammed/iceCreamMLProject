# PROBLEM STATEMENT
You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC). You decide that a Linear Regression model might be a good candidate to solve this problem.  
Data set:
- Independant variable X: Outside Air Temperature
- Dependant variable Y: Overall daily revenue generated in dollars

# STEP #1: LIBRARIES IMPORT


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# STEP #2: IMPORT DATASET

```python
IceCreamData = pd.read_csv("/content/IceCreamData.csv")
```

```python
IceCreamData.head(5)
```

```python
IceCreamData.describe()
```

# STEP#3: VISUALIZE DATASET

```python
sns.jointplot(x='Temperature', y='Revenue', data=IceCreamData, color='gray')
plt.show()
```

```python
sns.pairplot(IceCreamData)
plt.show()
```

```python
sns.lmplot(x='Temperature', y='Revenue', data=IceCreamData) #Showing  the best Line
```

# STEP#4: CREATE TESTING AND TRAINING DATASET

```python
X = IceCreamData.iloc[:, 0:1] #Temperature
y = IceCreamData.iloc[:, 1:] #Revenue
```

```python
from sklearn.model_selection import train_test_split
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

# STEP#5: TRAIN THE MODEL

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)
```

```python
# y = b + m*x
print("Linear Model Coeff (m):", model.coef_)
print("Linear Model Coeff (b):", model.intercept_)
```

# STEP#6: TEST THE MODEL

```python
#y_test = model.predict(X=x_test)
```

```python
plt.scatter(X_train, y=y_train, color="gray", )
plt.plot(X_train, model.predict(X_train), color="red")
plt.ylabel('Revenue [$]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temperature @Ice Cream Stand (Training)')
plt.show()
```

```python
plt.scatter(x_test, y=y_test, color="gray", )
plt.plot(x_test, model.predict(x_test), color="red")
plt.ylabel('Revenue [$]')
plt.xlabel('Temperature [degC]')
plt.title('Revenue Generated vs. Temperature @Ice Cream Stand (Testing)')
plt.show()
```

```python
model.score(x_test, y_test)
```
